```
%%writefile app.py
import streamlit as st
import streamlit as st
import numpy as np
from streamlit_lottie import st_lottie
from PIL import Image
import warnings

warnings.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import matplotlib as mpl
import requests
import streamlit as st
import streamlit.components.v1 as components
import hydralit_components as hc
import time
import cv2 as cv
import tempfile

############# face verfication model ####################
import pickle
import numpy
import tensorflow
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Layer, Conv2D, Dense, MaxPooling2D, Input, Flatten
import tensorflow as tf
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau , EarlyStopping
import warnings
warnings.filterwarnings('ignore')

model_path= '/content/drive/MyDrive/Automated Video Surveillance/face_verification/datasetV2/best_vgg_data_new_correct.h5'
vggface_path = '/content/drive/MyDrive/Automated Video Surveillance/face_verification/vgg_face_weights.h5'


####### build model ###
from tensorflow.keras.models import Model, load_model
from keras.applications import VGG16
from keras.layers import Lambda
from deepface.basemodels import VGGFace
import tensorflow as tf

def make_siamese_model():
    # Anchor image input in the network
    input_image = Input(name='input_img', shape=(224,224,3))
    
    # Validation image in the network 
    validation_image = Input(name='validation_img', shape=(224,224,3))

    vgg = VGGFace.baseModel()
    vgg.load_weights(vggface_path)
    cnt = 0
    for lyer in vgg.layers:
        if cnt > 34:
            lyer.trainable = True
            lyer.activation= tf.keras.layers.Activation('tanh')
        else:
            lyer.trainable = False
        cnt += 1
    vgg_face = Model(inputs=vgg.layers[0].input, outputs=vgg.layers[-2].output)

    x1 = vgg_face(input_image)
    x2 = vgg_face(validation_image)

    x1 = Flatten()(x1)
    x2 = Flatten()(x2)

    distances = Lambda(lambda x: tf.math.abs(x[0] - x[1]), name='L1_distance')([x1, x2])
    classifier = Dense(3, activation='softmax')(distances)
    
    siamese_model = Model(inputs=[input_image, validation_image], outputs=classifier)

    # Load the saved model with custom object scope
    with tf.keras.utils.custom_object_scope({'my_activation': my_activation}):
        saved_model_path = model_path
        loaded_model = tf.keras.models.load_model(saved_model_path)

    return loaded_model


import tensorflow as tf

def my_activation(x):
    return tf.keras.activations.tanh(x)

# Register the custom activation function with Keras
tf.keras.utils.get_custom_objects().update({'my_activation': my_activation})

# Define the saved model path
saved_model_path = model_path

# Load the saved model with the custom activation function
with tf.keras.utils.custom_object_scope({'Activation': tf.keras.layers.Activation,'my_activation': my_activation}):
  loaded_model = tf.keras.models.load_model(saved_model_path)

df = pd.read_csv('/content/drive/MyDrive/Automated Video Surveillance/face_verification/datasetV2/datasetv2.csv')

def preprocess(file_path):
    
  # Read in image from file path
  byte_img = tf.io.read_file(file_path)
  # Load in the image 
  img = tf.io.decode_jpeg(byte_img)
  
  # Preprocessing steps - resizing the image to be 100x100x3
  img = tf.image.resize(img, (224,224))
  # Scale image to be between 0 and 1 
  img = img / 255.0

  # Return image
  return img


import os
ver_s_path = '/content/drive/MyDrive/Automated Video Surveillance/face_verification/raw images/verify sarah'
ver_y_path = '/content/drive/MyDrive/Automated Video Surveillance/face_verification/raw images/verify yasmeen'

def verify(model, detection_threshold, verification_threshold, ver_s_path,ver_y_path,img_path):
  # Build results array
  results = []
  resulty = []
  ver1 = ver_s_path
  for image in os.listdir(ver1):
      input_img = preprocess(img_path)
      validation_img = preprocess(os.path.join(ver1, image))
      
      # Make Predictions 
      result = model.predict(list(np.expand_dims([input_img, validation_img], axis=1)))
      results.append(result)

  ver2 = ver_y_path
  for image in os.listdir(ver2):
      input_img = preprocess(img_path)
      validation_img = preprocess(os.path.join(ver2, image))
      
      # Make Predictions 
      result = model.predict(list(np.expand_dims([input_img, validation_img], axis=1)))
      resulty.append(result)
  ############## 
  other = 0
  sarah = 0
  yasmeen = 0

  for i in range(15):
    highest_index = np.argmax(results[i], axis=1)
    #print(highest_index)
    if highest_index == 0:
      other = other+1
    elif highest_index == 1:
      sarah = sarah+1
    elif highest_index==2:
      yasmeen = yasmeen +1

  for i in range(15):
    highest_index = np.argmax(resulty[i], axis=1)
    #print(highest_index)
    if highest_index == 0:
      other = other+1
    elif highest_index == 1:
      sarah = sarah+1
    elif highest_index==2:
      yasmeen = yasmeen +1

  variables = {'other': other, 'sarah': sarah, 'yasmeen': yasmeen}

  # find the key with the highest value
  max_value = max(variables, key=variables.get)
  return max_value
  





#######################################################

####################### stream lit app ################################


st.set_page_config(page_title="Automated Anomaly Detection", page_icon=":tada:", layout="wide")
def load_lottieurl(url):
    r = requests.get(url)
    if r.status_code != 200:
        return None
    return r.json()

# ---- LOAD ASSETS ----
lottie_coding = load_lottieurl("https://assets5.lottiefiles.com/packages/lf20_fcfjwiyb.json")

# ---- HEADER SECTION ----
with st.container():
    left_column, right_column = st.columns(2)
    with left_column:
        st.subheader("Face Verification and Anomaly Detection :wave:")
        #st.title("An Electronics and Communcation Engineer")
        st.write(
            "In this app we will deploy our project, let's get started :boom:"
        )
        #st.write("[Reach us >](https://www.linkedin.com/in/yassmen-youssef-48439a166/)")
    with right_column:
        st_lottie(lottie_coding, height=300, key="coding")



import streamlit.components.v1 as components
import hydralit_components as hc
import streamlit as st
from streamlit_option_menu import option_menu

options =['Face Verfication', 'Anomaly Detection']

site = st.sidebar.selectbox("Select one part ", options)

if st.sidebar.button('Start '):
  if site =="Face Verfication":
    with st.container():
        st.write("---")
        x = st.camera_input(label="Take a picture", key="camera_input_file")

        if x is not None:
          #with st.spinner('✨Now your picture is under processing'):
          #results = verify(loaded_model, 0.5, 0.5,ver_path, picture)
          #time.sleep(10)
          #st.image(x.getvalue(), caption="Your photo", use_column_width=True)
          captured_image = x.fromarray(camera.astype('uint8'), 'RGB')
          captured_image.save("/content/captured_image.jpg")
        with st.spinner('✨Now your picture is under processing'):
          result = verify(loaded_model, 0.5, 0.5,ver_s_path,ver_y_path, '/content/captured_image.jpg')
          time.sleep(10)
          print(result)
          st.write(result)

    

        

   ################### still under processing ####################
  if site =="Anomaly Detection":
      with st.container():
          st.write("---")
          f = st.file_uploader("Upload file")
          if f is not None:
            tfile = tempfile.NamedTemporaryFile(delete=False)
            tfile.write(f.read())
            vf = cv.VideoCapture(tfile.name)
            stframe = st.empty()
            while vf.isOpened():
                ret, frame = vf.read()
                # if frame is read correctly ret is True
                if not ret:
                    print("Can't receive frame (stream end?). Exiting ...")
                    break
                gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
                stframe.image(gray)



```
